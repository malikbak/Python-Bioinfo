```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ncbi.nlm.nih.gov/geo/query/acc.cgi?acc=GSE2109'

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find all td tags (table data cells)
    td_tags = soup.find_all('td')

    if td_tags:
        # Extract and print text from each td tag
        for td in td_tags:
            print(td.get_text())  # Print the text within each td tag
    else:
        print("No <td> tags found on the page.")
else:
    print(f"Failed to retrieve the page. Status code: {response.status_code}")
```

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ncbi.nlm.nih.gov/geo/query/acc.cgi?acc=GSE2109'

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find all td tags (table data cells)
    td_tags = soup.find_all('td')

    if td_tags:
        summary_found = False
        country_found = False
        for td in td_tags:
            text = td.get_text().strip()
            
            # Check if Summary text found
            if text == "Summary":
                summary_found = True
                continue
            
            # Extract text after Summary
            if summary_found:
                print("Text after Summary:", text)
                summary_found = False
            
            # Check if Country text found
            if text == "Country":
                country_found = True
                continue
            
            # Extract text after Country
            if country_found:
                print("Text after Country:", text)
                country_found = False
    else:
        print("No <td> tags found on the page.")
else:
    print(f"Failed to retrieve the page. Status code: {response.status_code}")
```
